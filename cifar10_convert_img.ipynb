```python
import pickle
import numpy as np
import os

DRIVE = '/content/drive/MyDrive/'

ML_PATH = "大伯 - ML_Code/cifar-10-batches-py"

if os.path.isdir(os.path.join(DRIVE, 'Colab Notebooks/')):
    PATH = os.path.join(DRIVE, 'Colab Notebooks/', ML_PATH)   # drive_1
else:
    PATH = os.path.join(DRIVE, ML_PATH)                       # drive_2

os.listdir(PATH)


def unpickle(file):
    with open(file, 'rb') as f:
        data = pickle.load(f, encoding = "bytes")
    return data

#convert a row vector into a picture
def convert_images(raw):
    image = raw[b'data']
    name = raw[b'filenames']
    image = image.reshape([-1, 3, 32, 32])
    image = image.transpose([0, 2, 3, 1])

    return image, name

data1 = unpickle(os.path.join(PATH, 'data_batch_1'))

print(type(data1))
print(data1.keys())
print(data1[b'data'].shape)

image, name = convert_images(data1)

import matplotlib.pyplot as plt

num = 104
plt.imshow(image[num])
plt.xlabel(name[num].decode('utf-8').split('_s')[0] + ' (' + str(num) + ')', fontsize = 16)
plt.show()
```
